```python
import os
import pandas as pd
import numpy as np
from sklearn.utils import shuffle

directory = r"/content/drive/MyDrive/processed"

label_mapping = {
    "takeoff": 0,
    "right": 1,
    "left": 2,
    "landing": 3,
    "forward": 4,
    "backward": 5
}

dfs_files = []


for root, dirs, files in os.walk(directory):
  for filename in files:
    if filename.lower().endswith(".csv"):
      file_path = os.path.join(root, filename)
      folder_name = os.path.basename(root).lower()


      if folder_name in label_mapping:
        df = pd.read_csv(file_path)
        df["label"] = label_mapping[folder_name]
        dfs_files.append(df)
      else:
        print(f"Unknown folder name: {folder_name}")

# concatenate all labeled into a datafram
if dfs_files:
  concatenated_df_files = pd.concat(dfs_files, ignore_index=True)

  # shuffle the concatenated_df_files dataframe
  concatenated_df_files = shuffle(concatenated_df_files, random_state=42).reset_index(drop=True)


  print(concatenated_df_files.head(100))
  print(concatenated_df_files.info())
  print(concatenated_df_files["label"].value_counts())
```

```python
X = concatenated_df_files.drop(columns=['label'])
Y = concatenated_df_files['label']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print(Y.value_counts())
```

```python

from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Create and train the model

model = RandomForestClassifier(n_estimators=25, random_state=42)
model.fit(X_train, Y_train)

#predict on test set
Y_pred = model.predict(X_test)

confusion_matrix = confusion_matrix(Y_test, Y_pred)

ACCURACY= accuracy_score(Y_test, Y_pred)
print(f"Accuracy: {ACCURACY:.4f}")

accuracy = accuracy_score(Y_test, Y_pred)
print(f"Accuracy: {accuracy}")
print(X_train.shape, X_test.shape)
print("confusion matrix")
print(confusion_matrix)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.heatmap(confusion_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=concatenated_df_files['label'].unique(), yticklabels=concatenated_df_files['label'].unique())
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.title('Confusion Matrix')
plt.savefig('confusion_matrix.png')
plt.show()
```
